```python
def construct_lat(s):
    if (s in lat_lng):
        return lat_lng[s]['lat']
    return 0
def construct_lng(s):
    if (s in lat_lng):
        return lat_lng[s]['lng']
    return 0
```

```python
ru['lat'] = ru['name'].apply(construct_lat)
ru['lng'] = ru['name'].apply(construct_lng)
```
