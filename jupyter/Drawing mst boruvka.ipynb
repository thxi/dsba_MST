```python
import pandas as pd
import numpy as np
import seaborn as sns
import folium
```

```python
edges = pd.read_csv('../data/mst_boruvka.csv')
cities = pd.read_csv('../data/ru_lat_lng_clear.csv')
```

```python
edges.head(2)
```

```python
coord = dict()
```

```python
for (_, city) in cities.iterrows():
#     coord[city['name']] = {'lat':city['lat'], 'lng':city['lng']}
    coord[city['name']] = (city['lat'], city['lng'])
```

```python
m = folium.Map(
    location=[61.254035, 73.396221],
    zoom_start=3,
    tiles='CartoDB positron'
)

used = dict()
for (_, edge) in edges.iterrows():
    city1, city2 = edge['node1'], edge['node2']
    p1, p2 = coord[city1], coord[city2]
    folium.PolyLine([p1, p2]).add_to(m)
    if (city1 not in used):
        used[city1] = 0
        folium.Marker([p1[0], p1[1]], popup=city1, icon=folium.Icon(icon='circle')).add_to(m)
    if (city2 not in used):
        used[city2] = 0
        folium.Marker([p2[0], p2[1]], popup=city2, icon=folium.Icon(icon='circle')).add_to(m)
    
m
```

```python
m.save('boruvka_map_markers.html')
```
