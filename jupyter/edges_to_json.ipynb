```python
import pandas as pd
```

```python
e = pd.read_csv('../data/mst_boruvka.csv')
e_json = list()
```

```python
cities = pd.read_csv('../data/ru_lat_lng_clear.csv')
coord = dict()
for (_, city) in cities.iterrows():
    coord[city['name']] = (city['lat'], city['lng'])
```

```python
for (_, edge) in e.iterrows():
#     p = {'city1': edge['node1'], 'city2' : edge['node2']}
    p = {'city1': {'name':edge['node1'], 'coord': coord[edge['node1']]}, \
         'city2': {'name':edge['node2'], 'coord': coord[edge['node2']]}}
    e_json.append(p)
```

```python
import json
```

```python
with open('../data/mst_boruvka.json', 'w') as f:
    json.dump(e_json, f)
```
